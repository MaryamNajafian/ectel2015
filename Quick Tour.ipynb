# Inline Plots

#### :: A Very Simple Plot

```python
import numpy as np
x = range(5)
y = [0,1,4,9,16]
plt.plot(x,y)
```

#### :: More Complex 

```python
x = np.linspace(2*(-np.pi),2*(np.pi), 100)
y1 = np.cos(x)
y2 = np.sin(x)
plt.plot(x,y1, marker='o', color='red')
plt.plot(x,y2, marker='o')
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2 # 0 to 15 point radiuses

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

# Pandas Data Analysis Library

#### Pandas is the core Python library for data science

- At the heart of Pandas is a data structure called a Data Frame
- A Data Frame is tabular in format, similar to a spreadsheet with rows and columns
- But a Data Frame also has row and column labels which also serve as indices for fast manipulation
- Data Frame is also optimized for data alignment, handling missing data, working with time series and much more
